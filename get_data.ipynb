# Flavour Analysis

Flavours can be (roughly) pinned down on a handful of aromatic compounds. Complementary pairings tend to exploit similarities in ingredients, either through their lineage (carrots and dill are good) or through the overlap of the compounds they produce (cinnamon and cloves both contain eugenol).

Can Wikipedia tell us what spices and herbs we should:
 * pair to make tasty food; and
 * substitute when we're missing some?

```r
library(jsonlite)
install.packages('curl')
```

```r
category.contents <- function(categories){
    
    lapply(categories, function(cat.q){fromJSON(paste0("https://en.wikipedia.org/w/api.php?action=query&list=categorymembers&cmtitle=",cat.q, "&cmtype=page&cmlimit=200&format=json"))$query$categorymembers})
}
(flavours <- category.contents("Category:Flavors")) 
```

```r
flavours$query$categorymembers
```

```r
(spices.herbs <- category.contents(c("Category:Herbs","Category:Spices")))
```
